```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np

# Number of coin tosses
n = 1000

# Probability of getting heads on a single toss
p = 0.5

# Possible number of heads (successes)
k_values = np.arange(0, n+1)

# Probability mass function for each k
probabilities = binom.pmf(k_values, n, p)

# Plot the binomial distribution curve
plt.bar(k_values, probabilities, color='blue', alpha=0.7)
plt.title('Binomial Distribution of Number of Heads in 1000 Coin Tosses')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```
